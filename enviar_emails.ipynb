```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders
from openpyxl import load_workbook
import os

caminho = r"C:\Users\caioz\Desktop\curso rpa\ListaEmail.xlsx"
planilha_origem = load_workbook(filename=caminho)
aba_origem = planilha_origem.active
linhas_total = aba_origem.max_row + 1

# O bloco 'with' abre a conexão e a mantém aberta para todo o processo
with smtplib.SMTP_SSL("smtp.gmail.com", 465) as smtp:
    # Lembre-se da recomendação de usar variáveis de ambiente para a senha no futuro!
    smtp.login("caioguilherme137@gmail.com", "ihzc lple tdde ywoy")

    # --- CORREÇÃO DE INDENTAÇÃO: O 'for' loop vai DENTRO do 'with' ---
    for x in range(2, linhas_total):
        email_destinatario = aba_origem[f'C{x}'].value
        nome = aba_origem[f'B{x}'].value

        # Ignora linhas que não tenham e-mail ou nome
        if not email_destinatario or not nome:
            continue # Pula para a próxima iteração do loop

        # Monta o pacote do e-mail
        msg = MIMEMultipart()
        # (Melhora Opcional) Personalizando o assunto e o corpo do e-mail
        msg["Subject"] = f"Relatório Personalizado para {nome}"
        msg["From"] = "caioguilherme137@gmail.com"
        msg["To"] = email_destinatario

        corpo_texto = f"Olá, {nome}.\n\nEste é um e-mail de teste com seu anexo personalizado.\n\nAtenciosamente."
        msg.attach(MIMEText(corpo_texto, "plain"))

        # Monta o nome e o caminho do anexo
        nome_do_arquivo = f"{nome}.xlsx"
        caminho_do_anexo = os.path.join(r'C:\Users\caioz\Desktop\curso rpa', nome_do_arquivo)
        
        # Verifica se o anexo existe
        if os.path.exists(caminho_do_anexo):
            with open(caminho_do_anexo, "rb") as attachment:
                part = MIMEBase("application", "octet-stream")
                part.set_payload(attachment.read())
            encoders.encode_base64(part)
            part.add_header("Content-Disposition", f"attachment; filename= {nome_do_arquivo}")
            msg.attach(part)
        else:
            print(f"AVISO: Anexo '{caminho_do_anexo}' não encontrado. O e-mail para {email_destinatario} será enviado sem anexo.")

        # Envia o pacote completo
        smtp.sendmail("caioguilherme137@gmail.com", email_destinatario, msg.as_string())
        
        print(f"E-mail para {email_destinatario} processado!")

print("Processo finalizado.")
```
